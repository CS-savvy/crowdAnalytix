## Importing csv file

```python
import csv
```

```python
with open('interview_scent.csv') as file:
    reader = list(csv.DictReader(file))
    
reader[:2]
    
```

 #### a) Find the total number of row in the csv file, ignoring the rows where `prediction` in "None".

```python
len(reader)
```

```python
NoOfRows = 0
for row in reader:
    if row['predictions'] == 'None': continue
    NoOfRows+=1

NoOfRows
```

#### b) Find the distinct values in the `prediction` column of csv.

```python
distinctValues = set() 

for row in reader:
    if row['predictions'] == 'None': continue
    
    row['predictions'] = row['predictions'].replace(" ","")
    words = row['predictions'].split(",")
    for x in words:
        distinctValues.add(x)
```

```python
distinctValues
```

#### c) Find the count of each distinct value appearing in the `prediction` column of csv.

```python
# preparing dictionary for the count

countDistinct = dict()
for x in distinctValues:
    countDistinct[x] = 0

len(countDistinct)

```

```python
# count of each words
for row in reader:
    if row['predictions'] == 'None': continue
    
    row['predictions'] = row['predictions'].replace(" ","")
    words = row['predictions'].split(",")
    for x in words:
        countDistinct[x]+=1
    
countDistinct
```

#### d) Submit the output of above question into csv file.

```python
column_names = ['Words','Count']

with open('result.csv', 'w') as result:
    writer = csv.DictWriter(result , fieldnames= column_names)
    writer.writeheader()
    for key,value in countDistinct.items():
        writer.writerow({'Words':key , 'Count':value})
        
```
